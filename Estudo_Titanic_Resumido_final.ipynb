```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold
from sklearn.model_selection import RepeatedKFold
from sklearn.linear_model import LogisticRegression
```

```python
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
def trata_sex(valor):
    if valor == 'female':
        return 1
    else:
        return 0
    
train['Sex_bin'] = train['Sex'].map(trata_sex)
test['Sex_bin'] = train['Sex'].map(trata_sex)
```

```python
#trata novas variaveis da train e da test

train['emb_s'] = (train['Embarked'] == 'S').astype(int)
train['emb_c'] = (train['Embarked'] == 'C').astype(int)
train['cab_nula'] = (train['Cabin'].isnull()).astype(int)
train['miss'] = (train['Name'].str.contains('Miss')).astype(int)
train['mrs'] = (train['Name'].str.contains('Mrs')).astype(int)
train['master'] = (train['Name'].str.contains('Master')).astype(int)
train['col'] = (train['Name'].str.contains('Col')).astype(int)
train['major'] = (train['Name'].str.contains('Major')).astype(int)
train['mr'] = (train['Name'].str.contains('Mr.')).astype(int)


test['emb_s'] = (test['Embarked'] == 'S').astype(int)
test['emb_c'] = (test['Embarked'] == 'C').astype(int)
test['cab_nula'] = (test['Cabin'].isnull()).astype(int)
test['miss'] = (test['Name'].str.contains('Miss')).astype(int)
test['mrs'] = (test['Name'].str.contains('Mrs')).astype(int)
test['master'] = (test['Name'].str.contains('Master')).astype(int)
test['col'] = (test['Name'].str.contains('Col')).astype(int)
test['major'] = (test['Name'].str.contains('Major')).astype(int)
test['mr'] = (test['Name'].str.contains('Mr.')).astype(int)

variaveis = ['Sex_bin','Age','Pclass','SibSp','Parch','Fare','emb_s','emb_c','cab_nula','miss','mrs','master','col','major','mr']

train.head()
```

```python

x = train[variaveis]
y = train['Survived']

x = x.fillna(-1) #substitui os missing por -1
```

```python
#testes para saber a assertividade

resultados = []

kf = RepeatedKFold(n_splits=2,n_repeats=10, random_state=10)

for linhas_treino, linhas_valid in kf.split(x):
    print('Treino:', linhas_treino.shape[0])
    print('Valid:', linhas_valid.shape[0])

    x_treino = x.iloc[linhas_treino]
    x_valid = x.iloc[linhas_valid]
    y_treino = y.iloc[linhas_treino]
    y_valid = y.iloc[linhas_valid]

    #mdelo = RandomForestClassifier(n_estimators=100,n_jobs=-1,random_state=0)
    modelo = LogisticRegression()
    modelo.fit(x_treino,y_treino)
    p = modelo.predict(x_valid)
    np.mean(y_valid == p)

    #armazenar essa média da acuraria em uma lista pra checar depois
    acc = np.mean(y_valid == p)
    resultados.append(acc)
print(np.mean(resultados))    
```

```python
#retreino final sem os split

modelo = LogisticRegression()
modelo.fit(x,y)
pf = modelo.predict(test[variaveis].fillna(-1))
pf
```

```python
#export

sub = pd.Series(pf,index=test['PassengerId'],name='Survived')
sub.to_csv('Modelo_Final.csv',header=True)
```
